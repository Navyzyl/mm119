### 取得每个籍贯对应的官员数量

### 取得每个籍贯对应的经纬度

### 将数据保存在json中

```python
address_dict = {}
address_dict['上元']=[118,32]
address_dict
```

```python
#方法
#1.把数据何存到json文件当中
#定义一个字典保存籍贯和对应的经纬度
address_dict = {}
for city in addr.index.tolist():
    address_dict[city]=addr.loc[city,['X','Y']].tolist()
address_dict
```

### 将字典保存到json文件中（将上面二合一）

```python
import json
with open('./addr1.json','w',encoding='utf-8') as f:
    json.dump(address_dict,f,ensure_ascii=False)
```

### 地理分布绘图

```python
#复制官网原形，然后更改相应设置
from pyecharts import options as opts
from pyecharts.charts import Geo
from pyecharts.faker import Faker

c = (
    Geo(init_opts=opts.InitOpts(width='1200px',height='800px'))
    .add_schema(maptype="china",is_roam=False)
    .add_coordinate_json('./addr1.json')
    .add("官员人数", [list(z) for z in zip(addr.index.tolist(), addr['官员人数'].values.tolist())])
    .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
    .set_global_opts(
        visualmap_opts=opts.VisualMapOpts(), title_opts=opts.TitleOpts(title="Geo-基本示例")
    )
#    .render("geo_base.html")
    .render_notebook()
)
c
```
