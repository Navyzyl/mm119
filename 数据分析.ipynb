#### 距离

```python
n['Distance 距离'].groupby(n['联系']).mean().sort_values()
#不同联系类型平均距离排序
```

```python
#绘图
n['Distance 距离'].groupby(n['联系']).mean().sort_values().plot(kind='bar')
plt.title('联系-平均距离值')
plt.xlabel('联系')
plt.ylabel('距离均值')
plt.show()
```

```python
#方法2
data=n['Distance 距离'].groupby(n['联系']).mean()
plt.barh(data.index,data.values)
plt.title('联系和距离关系',size=20,color='red')
plt.xlabel('平均距离')
plt.ylabel('联系')
plt.show()
```

```python
#支持性筛选
n.query("联系=='支持' or 联系=='政见趋同' or 联系=='得到Y的支持'")['Distance 距离'].mean()
```

```python
#对抗型筛选
n.query("联系=='不合' or 联系=='遭到Y的反对/攻讦' or 联系=='反对/攻讦'")['Distance 距离'].mean()
```
