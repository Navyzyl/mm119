```python
#不同联系的平均距离
data2=m[['联系','Distance 距离']].groupby('联系').agg({'Distance 距离':'mean'}).sort_values('Distance 距离',ascending=False)
data2
```

```python
from pylab import mpl
mpl.rcParams['font.sans-serif']=['SimHei']
#防止中文乱码
data2.plot(kind='bar')
plt.title('联系与距离关系')
plt.xlabel('联系')
plt.ylabel('距离')
plt.show()
```

```python
data2.plot(kind='barh')
plt.title('联系与距离关系',size=30,color='blue')
plt.xlabel('距离')
plt.ylabel('联系')
plt.show()
```

```python
#添加新列，指数年只差
m.columns
```

```python
m['指数年之差']=abs(m['Index year']-m['Assoc. Index Year'])
```

```python
m
```

```python
#筛选出联系人指数年存在的
m_new=m[m['Assoc. Index Year']>0]
m_new
```

```python
data3=m_new['指数年之差'].groupby(m_new['联系']).mean()
data3
```

```python
data3.plot(kind='bar')
plt.title('指数年之差与联系关系',size=30,color='blue')
plt.xlabel('联系')
plt.ylabel('指数年之差')
plt.show()
```

```python
x=data3.index
y=data3
plt.bar(x,y)
plt.title('指数年之差与联系关系',size=30,color='blue')
plt.xlabel('联系')
plt.ylabel('指数年之差')
plt.show()
```
