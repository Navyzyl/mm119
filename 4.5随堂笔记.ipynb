# df.apply方法

实例：添加一列温度类型：
1. 如果最高温度大于33度就是高温
2. 低于-10度是低温
3. 否则是常温

```python
#添加一个判断函数
def get_wendu_type(x):
    if x["bWendu"]>33:
        return '高温'
    if x["yWendu"]<-10:
        return '低温'
    return '常温'
```

```python
df.loc[:,"wendu_type"]=df.apply(get_wendu_type,axis=1)
df
```

1. 两只股票，是不是同涨同跌？程度多大？正相关还是负相关？

3. 协方差：衡量同向反向程度，如果协方差为正，说明x,y同向变化，协方差越大说明同向程度越高;
如果协方差为负，说明x,y反向运动，协方差越小说明反向程度越高

当相关系数为-1时，说明两个变量变化的反向相似度最大

在数据处理和分析中，merge和concat都是将不同的数据合并在一起的方法，但它们的实现方式和效果略有不同。

merge是将两个或多个数据集按照一定的条件进行连接，通常需要指定连接的键（key）或索引（index），它可以实现多种不同类型的连接，包括内连接、左连接、右连接和外连接等。merge通常用于将不同数据集中的信息按照某种关系进行关联，类似于SQL语句中的JOIN操作。

而concat则是将两个或多个数据集沿着某个轴（通常是行轴或列轴）进行简单的堆叠，它不需要指定连接的键或索引，只需确保合并的数据集具有相同的轴标签即可。concat通常用于将多个数据集拼接在一起，形成一个更大的数据集，类似于SQL语句中的UNION操作。

因此，merge和concat的主要区别在于它们的连接方式和合并方式不同。在具体应用中，需要根据数据集的特点和需求选择合适的方法来进行数据合并。
